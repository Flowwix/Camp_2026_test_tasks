```python
def solve_probability_task():
    # Data from task
    p_heads = [0.12, 0.27, 0.21, 0.96]
    priors = [0.25, 0.25, 0.25, 0.25]

    sequence = ['H', 'H', 'H', 'T', 'H', 'T', 'H', 'H', 'H']

    results = []

    # Iterative Update
    for flip in sequence:
        unnormalized_posteriors = []

        for i in range(4):
            if flip == 'H':
                likelihood = p_heads[i]
            else:
                likelihood = 1 - p_heads[i]

            # Posterior numerator = Likelihood * Prior
            val = likelihood * priors[i]
            unnormalized_posteriors.append(val)

        # Calculate Evidence (denominator) to normalize
        evidence = sum(unnormalized_posteriors)

        # Update priors for the next round (these become the current posteriors)
        priors = [x / evidence for x in unnormalized_posteriors]

        # Predict the NEXT flip
        # P(Next=H) = Sum( P(H|coin_i) * P(coin_i) )
        prob_next_h = sum(priors[i] * p_heads[i] for i in range(4))

        results.append(prob_next_h)

    # Output Formatted Results
    formatted_results = [f"{r:.2f}" for r in results]
    print(f"Probabilities after each step: {formatted_results}")

solve_probability_task()
```
